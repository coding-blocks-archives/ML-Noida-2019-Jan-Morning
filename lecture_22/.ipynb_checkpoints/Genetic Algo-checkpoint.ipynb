## Genetic Algo

 - Generate population
 - loop in
     1. check fitness
     2. cross
     3. mutate
 

#### Example of size 2 gnome

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1_range = [-100, 100]
x2_range = [-100, 100]
```

```python
def populate(features, size=1000):
    
    initial = []
    
    for _ in range(size):
        entity = []
        for feature in features:
            val = np.random.randint(*feature)
            entity.append(val)
        initial.append(entity)
        
    return np.array(initial)
```

```python
population = populate([x1_range, x2_range])

virus = np.array([5, 5])
```

```python
def draw(population, virus):
    plt.scatter(population[:, 0], population[:, 1], c='r', s=6)
    plt.scatter(virus[0], virus[1], c='g', s=100)
    plt.xlim([-100, 100])
    plt.ylim([-100, 100])
```

```python
draw(population, virus)
```

```python
def fitness(population, virus, size=100):
    
    scores = []
    
    for index, entity in enumerate(population):
        score = np.sum((entity - virus) ** 2)
        scores.append((score, index))
    
    scores = sorted(scores)[:size]
    
    return np.array(scores)[:, 1]
```

```python
def reduction(population, virus, size=100):
    
    fittest = fitness(population, virus, size)
    
    new_pop = []
    
    for item in fittest:
        new_pop.append(population[item])
    
    return np.array(new_pop)
```

```python
population = reduction(population, virus)
```

```python
draw(population, virus)
```
