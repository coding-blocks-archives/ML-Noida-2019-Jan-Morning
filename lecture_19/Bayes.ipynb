```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("../dataset/bays/mushrooms.csv")
```

```python
le = LabelEncoder()
```

```python
data = data.apply(le.fit_transform)
```

```python
def prior(df, col, label):
    return np.mean((df[col] == label).values)
```

```python
def cond_prob(df, col, label):
    return np.mean((df[col] == label).values)
```

```python
def classify(df, x_test, res_col):
    klasses = np.unique(df[res_col].values)
    probs = []
    
    for klass in klasses:
        pr = prior(df, res_col, klass)
        filtered = df[df[res_col]==klass]
        ll = 1
        
        for col in x_test.index.values:
            ll *= cond_prob(filtered, col, x_test[col])

        probs.append(pr*ll)
    
    return klasses[np.argmax(probs)]
            
            
```

```python
def score(df, res_col, X_test, y_test):
    
    y_pred = []
    
    X_test = X_test.drop([res_col], axis = 1)

    for index, row in X_test.iterrows():
        
        y_pred.append(classify(df, row, res_col))

#     return y_pred
    return (np.mean((y_test == y_pred).values))
    
```

```python
df_train, df_test = train_test_split(data, test_size=0.1)
```

```python
score(df_train, "type", df_test, df_test["type"])
```
