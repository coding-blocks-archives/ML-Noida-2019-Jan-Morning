```python
import numpy as np
```

```python
text = open("speeches/speech.txt", "r").read()

# text = "hello world hello world hello world hello world hello world "
```

```python
model = {}
```

```python
k = 10
```

```python
for i in range(len(text) - k - 1):
    X = text[i:i+k]
    y = text[i+k]
    if X not in model:
        model[X] = {}
    
    if y not in model[X]:
        model[X][y] = 1
    else:
        model[X][y] += 1
```

```python
sent = "We hope to join the"
```

```python
sent_cont = 0

while True:
    
    X_mod = sent[-k:]
    items = model[X_mod]
    
    if items is None:
        break
        
    vals = sorted(tuple(zip(items.values(), items.keys())))[-3 :]
    
    pred = vals[np.random.randint(0, len(vals))][1]
    
    sent += pred
    
    if pred == ".":
        sent_cont += 1
        
        if sent_cont == 10:
            break


print(sent)
```
